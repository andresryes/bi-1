```python
from sqlalchemy import create_engine
```

```python
source = create_engine('mysql+mysqlconnector://test:test123@bi-1_db_1/test')
```

```python
import pandas as pd

pd.read_sql('select current_date', con=source)
```

```python
import json
```

```python
with open('unconfirmed-transactions.json') as json_file:
    data = json.load(json_file)
```

```python
df_input = pd.json_normalize(data['txs'], record_path=['inputs'], meta=['lock_time', 'ver', 'size', 'weight', 'time', 'hash'])
df_input
```

```python
df_out = pd.json_normalize(data['txs'], record_path=['out'], meta=['lock_time', 'ver', 'size', 'weight', 'time', 'hash'])
df_out
```

```python
df_input['prev_out.spending_outpoints'] = df_input['prev_out.spending_outpoints'].astype(str)
```

```python
df_final = pd.merge(
    df_input,
    df_out,
    how="inner",
    on="hash",
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)

df_final
```
